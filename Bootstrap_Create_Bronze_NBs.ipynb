Run once to push the updated notebook (Step functions + Computed Columns).

```python
import requests, time
from notebookutils import mssparkutils

WORKSPACE_ID = "d445bd8c-ec46-44dd-96e1-ea6593a5b99b"
BASE = "https://api.fabric.microsoft.com/v1"
token = mssparkutils.credentials.getToken("https://api.fabric.microsoft.com")
hdr = {"Authorization": f"Bearer {token}", "Content-Type": "application/json"}

NOTEBOOKS = [
    {"displayName": "Bronze_BIX_LAH_DF2", "b64": ""},
]

r = requests.get(f"{BASE}/workspaces/{WORKSPACE_ID}/items?type=Notebook", headers=hdr, timeout=30)
r.raise_for_status()
existing = {i["displayName"]: i["id"] for i in r.json().get("value", [])}

def poll(loc, hdr, max_wait=120):
    for _ in range(max_wait//3):
        time.sleep(3)
        r = requests.get(loc, headers=hdr, timeout=20)
        if not r.ok: return None
        body = r.json()
        s = body.get("status","").lower()
        if s in ("succeeded","completed"): return body
        if s in ("failed","cancelled"): return None
    return None

for nb in NOTEBOOKS:
    name = nb["displayName"]
    defn = {"format":"ipynb","parts":[{"path":"notebook-content.ipynb","payload":nb["b64"],"payloadType":"InlineBase64"}]}
    if name in existing:
        item_id = existing[name]
        r = requests.post(f"{BASE}/workspaces/{WORKSPACE_ID}/items/{item_id}/updateDefinition", headers=hdr, json={"definition":defn}, timeout=90)
        if r.status_code == 202:
            loc = r.headers.get("Location") or r.headers.get("location")
            if loc: poll(loc, hdr)
        print(f"UPDATED: {name}  id={item_id}  status={r.status_code}")
    else:
        r = requests.post(f"{BASE}/workspaces/{WORKSPACE_ID}/items", headers=hdr, json={"displayName":name,"type":"Notebook","definition":defn}, timeout=90)
        item_id = "(async)"
        if r.status_code == 202:
            loc = r.headers.get("Location") or r.headers.get("location")
            if loc:
                res = poll(loc, hdr)
                if res: item_id = res.get("id", res.get("resourceId","ok"))
        elif r.status_code in (200,201):
            item_id = (r.json() or {}).get("id","?")
        print(f"CREATED: {name}  id={item_id}  status={r.status_code}")

print("Done.")
```
